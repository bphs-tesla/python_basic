# 5. 자료형(Type)이란? (2)

## 형 변환
### 암시적 변환

int와 float 사이에서는 암시적 변환 가능(float으로 변환됨)


### 명시적 변환

예를 들어, int와 str의 `합` 연산은 정의되지 않아 오류가 발생함.
```py
1 + "1" # TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

따라서 위를 처리해주기 위해서는 형 변환을 거쳐주어여함.

```py
1 + int("1") # 2
```


형 변환은 `int()` `float()` `str()` 등을 이용함.

## 묶음 자료형
### list
순서가 있는 자료형, 중복 가능, mutable
```py
array = [ 1, False, 'a' ]

print(array[0]) # 1
```

### tuple
순서가 있는 자료형, 중복 가능, immutable
```py

print(array[0])
```


### set (=집합)

```

### dict
딕셔너리, `키: 값` 형태
```py

# 6. 연산자 (2)

## 항등 연산자

- a `==` b (a=b?)
- a `!=` b (a≠b?)

-> 참/거짓(bool) 판단 (True/False)

## 대소 비교 연산자

- a `>` b (a>b?) 
- a `<` b (a<b?)
- a `>=` b (a≥b?)
- a `<=` b (a≤b?)

## 부정 연산자
- not a (~a)

* a가 truthy (참 같은 값) -> False
* a가 falsy (거짓 같은 값) -> True

```py
a = False
print(not a) # True
```

## and 연산자
`a and b`에서 조건 a 그리고 b가 모두 참일 때 `True`를 반환함.

cf) 다른 언어에서는 보편적으로 `&&` 연산자를 사용함.

| AND | T | F |
|-|-|-|
| T | T | F |
| F | F | F |

## or 연산자
`a or b`에서 조건 a 또는 b가 적어도 하나 참일 때 `True`를 반환함.

cf) 다른 언어에서는 보편적으로 `||` 연산자를 사용함.

| OR | T | F |
|-|-|-|
| T | T | F |
| F | F | F |

# 7. 조건문

프로그래밍에서 특정 코드를 조건이 만족하는 상황에서만 실행되도록 하는 제어 장치.


파이썬에서는 `if` `elif` `else`의 키워드를 사용함.

ex) 코드블럭(들여쓰기)이란?

````python
D = input("판별식의 값은?") # 입력값을 받고
D = float(D) # D의 형을 str에서 float으로 변환한다.

# D의 부호에 따라 조건문을 적용
if D > 0:
    print("두 개의 서로 다른 실근을 갖습니다.")

elif D == 0:
    print("한 개의 서로 다른 실근을 갖습니다.")

else:
    print("실근을 갖지 않습니다.")
````

# 8. 반목문

## range()
range(a, b)라면, a부터 b-1까지의 '숫자'가 담긴 리스트 같은? 객체를 반환한다~

## for문
for문은 Iterable한 객체(list, tuple, set, dict)에 저장된 데이터를 하나씩 가져와서 순회하는 반복문이다.

ex)
```py
for i in range(0, 5):
    print(i)
```
결과:
```
0
1
2
3
4
```


## while문
while문은 주어진 조건이 참일 때까지 반복하는 구문이다.

ex)
```py
i = 0
while i < 5:
    print(i)
    i = i + 1 # i+=1로도 축약 가능
```

결과:
```
0
1
2
3
4
```

## break
break 키워드를 사용하여 진행중인 반목문을 나갈 수 있다.

ex)
```py
i = 5
while True:
    print(i)
    i = i - 1 # i-=1로도 축약 가능
    if i == 3:
        print("Bye")
        break
```

결과:
```
5
4
Bye
```

## continue
continue 키워드를 사용하여 반목문에서 코드 실행을 건너뛸 수 있다.

ex)
```py
for i in range(100):
    if i % 3 == 1:
        continue
    print(i)
```

결과:
```
2
3
5
6
...
96
98
99
```
